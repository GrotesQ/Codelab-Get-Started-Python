```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time

browser = webdriver.Chrome( '/Users/unk/tools/chromedriver' )
browser.get( 'https://www.tpirates.com' )
time.sleep( 2 )
browser.find_element_by_link_text( '시장' ).click()
time.sleep( 2 )
browser.find_element_by_id( 'searchZoneCode' ).click()
browser.find_element_by_class_name( 'targetItem0000000002_0' ).click()
browser.find_element_by_id( 'sorting' ).click()
browser.find_element_by_class_name( 'targetItempopularity' ).click()
time.sleep( 1 )

html = BeautifulSoup( browser.page_source, 'html.parser' )
cards = html.find( id='searchList' ).find_all( class_='card-template' )
links = []

for card in cards :
    link = card.find(class_='use-hyperlink')[ 'data-href' ]
    links.append( link )

for link in links :
    browser.get( link )
    time.sleep( 2 )
    html = BeautifulSoup( browser.page_source, 'html.parser' )
    name = html.find( class_='name' ).get_text().strip()
    call1 = html.find( class_='use-call-count' ).get_text().strip()
    print( name + ' / ' + call1 )

browser.close()
```
