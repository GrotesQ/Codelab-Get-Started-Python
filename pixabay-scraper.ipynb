```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import urllib.request

keyword = input('검색어를 입력하세요')
browser = webdriver.Chrome('/Users/unk/tools/chromedriver')
browser.get( 'https://pixabay.com/ko/images/search/' + keyword + '/' )

html = BeautifulSoup( browser.page_source, 'html.parser' )
items = html.find( class_='search_results' ).find_all( class_='item' )
paths = []

for item in items :
    paths.append( item.find( 'img' )['src'] )

ua = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.97 Safari/537.36'
opener = urllib.request.build_opener()
opener.addHeaders = [ ('User-Agent', ua ) ]
for path in paths :
    try :
        urllib.request.install_opener( opener )
        urllib.request.urlretrieve( path )
    except :
        print( '로드 실패 : ' + path )

browser.close()
```
